```python
import numpy as np
import sys
import simpy
```

```python
class Node:
    def __init__(self, env, name):
        self.env = env
        self.name = name
        self.addr = name
        self.links_in = []
        self.links_out = []
        self.output_port = []
        self.routing_table = {}
        
class Switch(Node):
    def __init__(self, env, name):
        super().__init__(env, name)
        
class Host(Node):
    def __init__(self, env, name, applications=None):
        super().__init__(env, name)
        self.applications = applications if applications is not None else []
        self.queue = simpy.Store(env)
        self.delay_dict = {}
```

```python
env = simpy.Environment()
print(env)
node1 = Host(env, 1)
node2 = Host(env, 2)
node3 = Switch(env, 3)
node4 = Switch(env, 4)
```

```python
class Topology:
    def __init__(self, env):
        self.env = env
        #self.topology_table = topology_table
        self.nodes = {}
        self.links = {}
```

```python
topo = Topology(env)

if node1 not in topo.nodes:
    topo.nodes.update({node1 : node1})
    topo.links.update({node1 : {}})
    
if node2 not in topo.nodes:
    topo.nodes.update({node2 : node2})
    topo.links.update({node2 : {}})
    
if node3 not in topo.nodes:
    topo.nodes.update({node3 : node3})
    topo.links.update({node3 : {}})
    
if node4 not in topo.nodes:
    topo.nodes.update({node4 : node4})
    topo.links.update({node4 : {}})
```

```python
class Link(simpy.Store):
    #링크
    #필요 변수: 용량(capacity, bps단위), 물리적 거리(m단위)
    def __init__(self, env, link_capacity=100000000, distance=10):
        super().__init__(env)
        self.link_capacity = link_capacity
        self.distance = distance
```

```python
def add_link(node10, node20):
    if node10 not in topo.nodes:
        topo.nodes.update({node10 : node10})
        topo.links.update({node10 : {}})
    if node20 not in topo.nodes:
        topo.nodes.update({node20 : node20})
        topo.links.update({node20 : {}})
    if node20 not in topo.links[node10]:
        link1 = Link(env)
        link2 = Link(env)
        topo.links[node10][node20] = link1
        topo.links[node20][node10] = link2
        topo.nodes[node10].add_link(link1, link2)
        topo.nodes[node20].add_link(link2, link1)
```

```python
add_link(node1, node3)
add_link(node3, node4)
add_link(node4, node2)
```
